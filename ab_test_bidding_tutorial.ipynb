# AB Testi (Bağımsız İki Örneklem T Testi)

Bu projede AB Testi kullanılarak testin nihai sonuçlarının analizi yorumlanacaktır. Proje detaylarına girmeden önce AB testinden, kullanma amacımızdan ve izleyeceğimiz yoldan bahsedip kodlara geçiş yapılacaktır.

## AB Testinin Kullanım Amacı Nedir?
"Bir test yapıyoruz ama bu testimizin amacı ne diye?" soracak olursak temel amacımız olası farklılıkların şans eseri ortaya çıkıp çıkmadığını göstermeye çalışmaktır. Örneğin şirket, bir makine öğrenmesi projesine yatırım yapmış ve ürettiği makine öğrenmesi modelleri ile kullanıcılarına çeşitli ürün tavsiyelerinde bulunuyor olsun. Öyle ki elimizde önceden makine öğrenmesi kullanılmayan eski sistemimiz vardı, bir de emek ve para harcanıp geliştirilen makine öğrenmesi projesi neticesinde kullanıcılara ürün önerilerinin akıllı bir şekilde yapıldığı yeni sistemimiz var. Peki bu yapılan bu yatırım gerçekten emeğimize, paramıza değiyor ve işe yarıyor mu? İlk bakışta matematiksel olarak fark gözükse de, bu fark gerçekten var mı, yoksa <ins>şans eseri mi ortaya çıktı?</ins> Başta da dediğimiz gibi aslında AB testinin amacının anahtar özeti de budur: <ins>Karşılaştırmalarında temel amaç, olası farklılıkların şans eseri ortaya çıkıp çıkmadığını göstermeye çalışmaktır.</ins>

Dikkat edilmesi gereken nokta, burada geliştirilen makine öğrenmesi modelinin başarısıyla ilgilenmiyoruz. İlgilendiğimiz şey, bu makine öğrenmesi modeli sisteme entegre edildiğinde ve bunun önerilerine göre kullanıcılar satın alma işlemlerini gerçekleştirdiğinde ortaya çıkacak olan gelirlerin, eski sisteme göre anlamlı ve istatistiki bir farklılık yaratıp yaratmadığıdır. 


## AB Testi Nedir ve Nasıl Uygulanır?
İki grubun ortalamaları veya oranları arasında karşılaştırma yapılmak istenildiğinde kullanılır. 

A kontrol grubunu, B deney grubunu temsil eder.

### 1. Hipotez Kurma
**Hipotez:** Bir inanışı, bir savı test etmek için kullanılan istatistiksel yöntemdir.



Buradaki H0, yokluk hipotezidir ve bizim sınayacağımız durumdur. "İki grubun ortalamaları arasında fark yoktur", bir başka deyişle "İki grubun ortalamaları eşittir" hipotezini kurup bu hipotezi sınarız. Sınama sonucunda hipotezi reddetme veya reddedememeye göre iki grup ortalaması arasında fark olup olmadığını değerlendiririz.

Yapacağımız işlemlerde ortalamaların farklarını zaten biliyoruz, reddedecek olduğumuzda farkın hangisinden kaynaklandığını bilip buna göre yorum yapabilmek adına yukarıdaki karşılaştırmalardan en soldakini kullanacağız.

<ins>Özetle, testlerin sonucunda iki grubun ortalamalarını kıyaslayacağız ve bunların arasında "fark var" veya "fark yok" diyeceğiz.</ins>

### 2. Varsayım Kontrolü

Kullanacağımız, bize bilimsel bazı kıyaslamalar ve karara varma imkanları sağlayan yöntemlerin bazı varsayımları vardır. Bağımsız İki Örneklem T Testinin 2 varsayımı vardır:
* <ins>Normallik Varsayımı:</ins> İki grubun da ayrı ayrı standart normal dağılması gerekmesi varsayımıdır. Shapiro-Wilks Testi kullanılır. 
* <ins>Varyans Homojenliği:</ins> İki grubun varyanslarının homojenliği varsayımıdır. Yani iki grubun dağılımlarının birbirine benzer olup olmamasına bakılır. Levene Testi kullanılır. 

**Varsayım kontrollerini neden yapıyoruz?:** Çünkü varsayım kontrolleri sonda yapacağımız bağımsız iki örneklem T testinin "parametrik" mi yoksa "non-parametrik" mi olacağını söyler. Varsayımlarımız için de "H0: Varsayım sağlanıyordur" şeklinde bir hipotez kurup yaptığımız testin sonucuna göre parametrik veya non-parametrik test yaparız. 

<ins>Yani test içinde test mi çıktı?:</ins> Cevabımız evet. Varsayım testlerinin amacı sonda yapacağımız testin nasıl uygulanacağını belirlemek içindir. Varsayımlar içinde kendi içinde "H0: Varsayım sağlanıyordur" hipotezleri kurulur, bunların reddedilmesine veya reddedilmemesi T testini nasıl yapacağımıza dair bize yol gösterir.

## 3. Testin Uygulanması

Yukarıda da belirttiğimiz gibi burada uygulayacağımız test varsayım kontrollerimize göre şekillenir. Burada varsayım kontrollerinin reddedilmesine veya reddedilmemesine göre farklı yollar karşımıza çıkar:
* Eğer iki varsayım da sağlanıyorsa parametrik testimiz olan "bağımsız iki örneklem t testi" uygulanır.
* Normallik sağlanıyor fakat varyans homojenliği sağlanmıyorsa yine parametrik testimiz "bağımsız iki örneklem t testi" uygulanır fakat bu durumda testin uygulanacağı fonksiyona varyans homojenliğinin sağlanmadığını belirten argüman eklenir.
* Normallik sağlanmıyorsa direkt olarak non-parametrik testimiz olan "mannwhitneyu testi" uygulanır. 

## 4. Bağımsız İki Örneklem T Testi ve Yorumlanması
Testler için ilgili fonksiyonları kullanıp hipotez testlerini gerçekleştirdiğimizde, ilgili fonksiyonları kullandığımızda bu fonksiyonlar bize bir **p-value** değeri veriyor olacak. Kurduğumuz hipotezin sonucunu testlerin sonucunda elde edeceğimiz p-value değerine bakarak yorumlayacağız. Bu p-value değeri **0.05’ten küçükse H0 reddedilir** diyerek çalışmanın başında kurduğumuz H0 her neyse, ona göre yorumlarımızı yapacağız.



# ÖZET
Kaybolmamak adına uygulacağımız adımların özetine tekrar bakalım.
1. Hipotez kur. 
> H0: Karşılaştırdığımız iki ortalama arasında istatistiki olarak anlamlı bir farklılık vardır.<br> 
H1: Karşılaştırdığımız iki ortalama arasında istatistiki olarak anlamlı bir farklılık yoktur.
2. Varsayım testlerini yaparak varsayımları incele.
> * Normallik Varsayımı: Shapiro Testi <br> 
> * Varyans Homojenliği: Levene Testi <br> 
> Ayrı ayrı test sonuçlarını incele, eğer p-value değeri 0.05'ten küçük olan varsayım reddedildi demektir, yani p-value değeri 0.05'ten küçük olan varsayım sağlanmamıştır. 
3. T testini uygula, p-value'ya bakarak final yorumunu yap.
> T testinin nasıl yapılacağı varsayımların reddedilme durumlarına göre şekillenir. <br>
>* İki varsayım da sağlanıyorsa: Parametrik T Testi <br>
>* Normallik sağlanıyor, varyans sağlanmıyorsa: Parametrik T Testi fakat varyansın sağlanmadığı argümanla belirlir.<br>
>* Normallik sağlanmıyorsa: Non-parametrik Mannwhitneyu Testi <br>
> Uyguladığımız testin sonucundaki p-value 0.05'ten küçükse başta kurduğumuz hipotez REDDEDİLİR, yani "Karşılaştırdığımız iki ortalama arasında istatistiki olarak anlamlı bir farklılık vardır" hipotezi reddedildi ve fark olduğu sonucuna varıldı. 

**EK BİLGİ:** <ins>H0'I YA REDDEDERİZ YA REDDEDEMEYİZ, H1'İ KABUL ETMEK DİYE BİR YORUM YOKTUR!</ins><br>
İstatistik literatüründe hipotez kurulduğunda hipotezi(H0'ı) reddederiz ya da reddedemeyiz. H1'i kabul etmek 
veya reddetmekle ilgili bir yorumumuz OLAMAZ çünkü H0'ı reddettiğimizde yapacak olduğumuz hata miktarını, yani p-value değerimizi biliriz(0.05). Ama H1'i kabul etmekle yapacak olduğumuz hatayı bilemeyiz, bir yere dayandıramayız.

```python
# Veri setlerini birleştirelim
df_final = pd.concat([df_control, df_test], axis=1)
df_final.head()
```

## 1. Hipotez

* **H0: M1 = M2** 
* **H1: M1 != M2**

Buradaki hipotezlerimizin anlamlarını tekrar hatırlayacak olursak:
* **H0:** Average bidding ve Max. bidding'in Purchase ortalamaları arasında istatistiki olarak anlamlı bir fark yoktur.
* **H1:** Average bidding ve Max. bidding'in Purchase ortalamaları arasında istatistiki olarak anlamlı bir fark vardır.

## 2. Varsayım Kontrolleri
Kontrol edeceğimiz 2 varsayım(Normallik Varsayımı, Varyans Homojenliği) ve bunlara uygulayacağımız testler vardı.

Burada yapmamız gereken şey 2 varsayım için de ayrı ayrı hipotez kurmak. Temelde yaptığımız şeyse "Varsayım sağlanıyor" hipotezini kurup reddedip reddetmediğimize bakmak

### Normallik Varsayımı (Shapiro-Wilks Testi)
* **H0:** Normallik varsayımı sağlanmaktadır.
* **H1:** Normallik varsayımı sağlanmamaktadır.


```python
test_stat, pvalue = shapiro(df_final["Purchase_control"])
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

```python
test_stat, pvalue = shapiro(df_final["Purchase_test"])
print('Test Stat = %.4f, p-value = %.4f' % (test_stat, pvalue))
```

Yukarıda görebileceğimiz üzere karşılaştıracağımız iki grubun da p-value değeri 0.05'ten yüksek. Bir hipotezi eğer p-value değeri 0.05'ten düşük olursa reddediyorduk, aksi durumda reddedemiyorduk. Yani burada iki grup için de diyebiliriz ki **p-value değeri 0.05'ten yüksek olduğu için H0 hipotezi REDDEDİLEMEZ.** Normallik varsayımı için H0 hipotezimiz "Normallik varsayımı sağlanmaktadır" şeklindeydi, biz bu hipotezi p-value sonucunda reddetmedik, yani normallik varsayımı sağlanmış oldu. 

### Varyans Homojenliği (Levene Testi)
* **H0:** Varyanslar homojendir.
* **H1:** Varyanslar homojen değildir.

```python
test_stat, pvalue = levene(df_final["Purchase_control"],
                           df_final["Purchase_test"])
print("Test stat = %.4f, p-value = %.4f" % (test_stat, pvalue))
```

Yukarıdakine benzer bir şekilde iki grubun da p-value değeri 0.05'ten yüksek. Aynı şekilde burada iki grup için de diyebiliriz ki **p-value değeri 0.05'ten yüksek olduğu için H0 hipotezi REDDEDİLEMEZ.** Varyans homojenliği için H0 hipotezimiz "Varyanslar homojendir" şeklindeydi, biz bu hipotezi p-value sonucunda reddetmedik, yani varyans homojenliği sağlanmış oldu. 

## 3. Testin Uygulanması ve Yorumu

Burada uygulayacağımız testi varsayım kontrollerinin sonucunun belirleyeceğinden bahsetmiştik. Başlamadan önce hangi durumda hangi testi uygulayacağımızı hatırlayalım:
* İki varsayım da sağlanıyorsa: Parametrik T Testi
* Normallik sağlanıyor, varyans sağlanmıyorsa: Parametrik T Testi fakat varyansın sağlanmadığı argümanla belirlir.
* Normallik sağlanmıyorsa: Non-parametrik Mannwhitneyu Testi

İki varsayım hipotezleri de reddedilemedi, yani varsayımlar sağlandı. O halde biz doğrudan parametrik T testimizi uygulayıp sonucumuza bakabiliriz.

```python
# eğer normallik sağlanıp varyans homojenliği sağlanmasaydı equal_var=False girilirdi
test_stat, pvalue = ttest_ind(df_final["Purchase_control"],
                              df_final["Purchase_test"],
                              equal_var=True)
print("Test stat = %.4f, p-value = %.4f" % (test_stat, pvalue))
```

Son olarak asıl başta kurduğumuz hipotez için p-value değeri 0.3493 olarak bulundu. Bu da demektir ki **p-value=0.3493 > 0.05 olduğundan dolayı H0 hipotezi REDDEDİLEMEZ.** Başta kurduğumuz hipotezi hatırlayalım:
* **H0:** Average bidding ve Max. bidding'in Purchase ortalamaları arasında istatistiki olarak anlamlı bir fark yoktur.

Biz H0 hipotezini REDDEDEMEDİK, bu demektir ki testten önce projenin başında Purchase ortalamalarında bir fark görmüştük fakat bu farklılık <ins>şans eseri sonucu ortaya çıktığı ve istatistiki olarak bir fark sağlanamamaktadır.</ins> "Average Bidding" sisteminin sonucunda önceki "Maximum Bidding" sisteminden Purchase anlamında farkı olmadığını ve bir katma değer uygulamadığı sonucuna varabiliriz. Dolayısıyla müşterimize buraya harcama ve yatırım yapmasını durdurup eski sistemden devam etmesinde bir sakınca olmadığı yönünde tavsiye verebiliriz.
